```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=np.nan,strategy="mean")
df.iloc[:,1:3]=si.fit_transform(df.iloc[:,1:3])
```

# -  Use OneHotEncoder to Features 



```python
background = pd.get_dummies(x['Background'])
```

```python
df = pd.concat([background,df],axis=1)
```

```python
df
```

```python
df.drop('Background',axis=1,inplace=True)
```

```python
df
```

## - Use LabelEncoder to Target



```python
y = df[["Remark"]]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y[["Remarks"]]= le.fit_transform(y[["Remark"]])
```

```python
df = pd.concat([df,y],axis=1)
```

```python
df
```

```python
df.drop('Remark',axis=1,inplace=True)
```

```python
df
```

# Do Feature Scaling with StandardScaler

```python
from sklearn.preprocessing import StandardScaler
for col in df:
    ss = StandardScaler()
    df[col]=ss.fit_transform(df[[col]])
```

```python
df
```

```python
df.info()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

# - Split the Training and Testing set



```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(xtrain,ytrain)
ypred = linreg.predict(xtest)
```

# Checking Accuracy

```python
from sklearn.metrics import mean_absolute_error as mae , mean_squared_error as mse , r2_score
```

```python
print(f"MAE :- {mae(ytest,ypred)}")
print(f"MSE :- {mse(ytest,ypred)}")
print(f"RMSE :- {mse(ytest,ypred)**0.5}")
print(f"Accuracy :- {r2_score(ytest,ypred)}")
```
